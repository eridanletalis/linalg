__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$
Решение:

\begin{cases}
  f_3(x) = x + f_2(x) \\
  f_4(x) = x - f_1(x)
\end{cases}



Вычтем из первого уравнения второе:
$$f_3(x) - f_4(x) = f_1(x) - f_2(x)$$

Векторы линейно зависимы, т.к можно из одного вектора выразить другие.

$$\alpha_1{2} + \alpha_2{x} + \alpha_3{x^2} + \alpha_4{x^2}  + \alpha_4{2x} + \alpha_4$$

Из системы видно, что все четыре вектора не будут линейно зависимыми, так как система из $\alpha_n$ не имеет решений, кроме тривиального.

5. Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;

б) все векторы, являющиеся линейными комбинациями данных векторов $${u1,u2,...,u_n}$$.

Решение:

__а.__) Не является, т.к. при выполнении операции сложения мы можем выйти из подпространства с этими условиями, например, возьмём три вектора у которых одна координата равна нулю: $(0, b_1, c_1), (a_2, 0, c_2), (a_3, b_3, 0), a_n,b_n,c_n $ не равны нулю. Сложим эти векторы покоординатно. Результатом сложения будет вектор $(a_2+a_3, b_1+b_3, c_1+c_2)$ у которого все три координаты не равны нулю, что противоречит условию этого подпространства.

__б.__) Возьмём два вектора $a = (\alpha_1{u_1} + \alpha_2{u_2} + ... + \alpha_n{u_n})$ и $b = (\beta_1{u_1} + \beta_2{u_2} + ... + \beta_n{u_n})$, оба векторая являются линейными комбинациями исходных векторов. Проверим, что операции суммы векторов и умножения на число не выводят результат за подпространство. $a+b = (\alpha_1 + \beta_1)u_1 + (\alpha_2 + \beta_2)u_2 + ... + (\alpha_n + \beta_n)u_n$. Видим, что результирующий вектор также находится в исходном подпространстве. Проверим умнижение на число:  $$\lambda{a} = (\lambda\alpha_1{u_1} + \lambda\alpha_2{u_2} + ... + \lambda\alpha_n{u_n}) = \lambda(\alpha_1{u_1} + \alpha_2{u_2} + ... + \alpha_n{u_n}). $$ Также видно, что умножение на число не выводит нас из исходного подпространства. 
Таким образом показано, что подпространство, состоящее из линейных комбинаций векторов линейного пространства являетя линейным.

__2.__ Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.


Формула для p-нормы (норма Гёльдера) ля вектора $x = (x_{1}, \dots, x_{n}) \in \mathbb{R}^{n}$

$$\left\| x \right\|_{p} = \left( \sum_{i} \left| x_{i} \right|^{p} \right)^{1 / p},~p \geq 1.$$

В нашем случае: $$||x_1||_{p} = (4^p + 2^p + 4^p)^{1/p}; ||x_1|| = 10, ||x_1||_{2} = 6$$
$$||x_2||_{p} = (12^p + 3^p + 4^p)^{1/p}; ||x_2|| = 19, ||x_2||_{2} = 13$$

Угол рассчитывается, как $$cos\varphi = \frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|}.$$, где нормы - евклидовские ($p_2$) 




```
from numpy.linalg import norm

a = np.array([4, 2, 4])
b = np.array([12, 3, 4])
cos_phi = np.dot(a, b) / norm(a) / norm(b)
print(f"Косинус угла между a и b: {cos_phi:.2f}")
print(f"Угол между a и b: {np.arccos(cos_phi):.2f} радиан или {np.arccos(cos_phi) * 180 / np.pi:.2f} градусов.")
print(f"p-1 норма вектора а {norm(a,1)}, p-2 норма вектора а {norm(a,2)}")
print(f"p-1 норма вектора а {norm(b,1)}, p-2 норма вектора b {norm(b,2)}")
```

__4.__ Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$


Работаем с определением
__Определение.__ В конечномерном евклидовом пространстве базис $\{e_{1},e_{2},...,e_{n}\}$ называется ортонормированным, если 

$(e_{i}, e_{j})=0$ $\forall$ $i\neq j$ и $(e_{i},e_{i})=1$ $\forall$ $i\in[1, n].$

__а.__) не образуют, нет третьего вектора, а измерение трёхмерное (если третий нулевой, то его скалярное произведение само на себя не равно 1)<br/>
__б.__) образуют, т.к. скаляры самих на себя равны единице, а комбинации нулевые <br/>
__в.__) не образуют, т.к. $(\frac{1}{2},-\frac{1}{2},0)(\frac{1}{2},-\frac{1}{2},0) = \frac{1}{4} + \frac{1}{4} + 0 \neq 1$ <br/>
__г.__) образуют, являются стандартным базисом трёхмерного пространства.
